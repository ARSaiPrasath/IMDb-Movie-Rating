# Cleaning the Dataset

```python
Inspecting Null values
```

```python
#code for column-wise null count
movies.isnull().sum()
```

```python
# code for row-wise null count
movies.isnull().sum(axis=1)
```

```python
# code for column-wise null percentages 
a = movies.isnull().sum()/len(movies)*100
print(a)
b=[]
for items in a.iteritems():
  if (items[1]>5):
    b.append(items[0])
b
```

Dropping unecessary columns

```python
#  code for dropping the columns here
columns = ["color","director_facebook_likes","actor_1_facebook_likes","actor_2_facebook_likes","actor_3_facebook_likes","actor_2_name","cast_total_facebook_likes","actor_3_name","duration","facenumber_in_poster","content_rating","country","movie_imdb_link","aspect_ratio","plot_keywords"]
movies = movies.drop(columns,axis=1)
movies
```

Dropping unecessary rows using columns with high Null percentages

```python
# code for dropping the rows 
for i in columns:
  if(i in b):
    b.remove(i)
movies = movies.dropna(subset=b)
movies

```

Filling NaN values

```python
# code for filling the NaN values in the 'language' column 
movies["language"] = movies["language"].fillna(value="English")
movies
```

Checking the number of retained rows

```python
# code for checking number of retained rows 
movies.isnull().sum()
```

**Checkpoint 1:** You might have noticed that we still have around `77%` of the rows!

# Data Analysis

Change the unit of the `budget` and `gross` columns from `$` to `million $`

```python
# code for unit conversion 
collection_tmp = movies["gross"]
collection_bdgt = movies["budget"]
millions1 = pd.to_numeric(collection_tmp, errors='coerce')/1e6
millions2 = pd.to_numeric(collection_bdgt, errors='coerce')/1e6
movies['gross'] = millions1.round(2).astype('str')
movies['budget'] = millions2.round(2).astype('str')
movies
```

Finding the movies with highest profit

```python
#code for creating the profit column 
movies['profit'] = movies['gross'].astype(float)-movies['budget'].astype(float)
movies
```

```python
# code for sorting the dataframe 
movies=movies.sort_values(by="profit",ascending=False)
movies
```

```python
# code for profit vs budget plot 
import matplotlib.pyplot as plt
x = movies['profit']
y = movies['budget']
plt.plot(x,y)
plt.show()
```

```python
# code to get the top 10 profiting movies 
top10 = movies.iloc[:10]
top10
```

Finding popular genres

```python
# code for extracting the first two genres of each movie
movies['genres']=movies.genres.str.split('|')
movies['genre_1']=movies['genres'].apply(lambda x: x[0])  
movies['genre_2']=movies['genres'].apply(lambda x: x[1] if len(x)>1 else x[0])   
movies
```

```python
# code for grouping the dataframe 
movies_by_segment = movies.groupby(['genre_1','genre_2'])
movies_by_segment
```

```python
# Write your code for getting the 5 most popular combo of genres here
PropGenre = pd.DataFrame(movies_by_segment.gross.mean().sort_values(ascending=False) ) 
PropGenre[0:5]
```

**Checkpoint 5:** Well, as it turns out. `Family + Sci-Fi` is the most popular combo of genres out there!

Find the critic-favorite and audience-favorite actors

```python
# code for creating three new dataframes 
Meryl_Streep = movies[['actor_1_name','movie_title','num_critic_for_reviews','num_user_for_reviews']]
Leo_Caprio = movies[['actor_1_name','movie_title','num_critic_for_reviews','num_user_for_reviews']]
Brad_Pitt = movies[['actor_1_name','movie_title','num_critic_for_reviews','num_user_for_reviews']]
# Include all movies in which Meryl_Streep is the lead
Meryl_Streep=Meryl_Streep.loc[Meryl_Streep['actor_1_name']=='Meryl Streep',:]
Meryl_Streep.head()
```

```python
# Include all movies in which Leo_Caprio is the lead
Leo_Caprio=Leo_Caprio.loc[Leo_Caprio['actor_1_name']=='Leonardo DiCaprio',:]
Leo_Caprio.head()
```

```python
# Include all movies in which Brad_Pitt is the lead
Brad_Pitt=Brad_Pitt.loc[Brad_Pitt['actor_1_name']=='Brad Pitt',:]
Brad_Pitt.head()
```

```python
# code for combining the three dataframes 
Combi=Meryl_Streep.append(Leo_Caprio).append(Brad_Pitt)
Combi
```

```python
# Write your code for grouping the combined dataframe here
Actor_name=Combi.groupby('actor_1_name')
Actor_name
```

```python
# code for finding the mean of critic reviews and audience reviews 
Critic_reviews=Actor_name['num_critic_for_reviews'].mean().sort_values(ascending=False)
Critic_reviews
```

**Checkpoint 6:** `Leonardo` has aced both the lists!
